# DBMZ bertle training


```
from transformers import  TFBertModel, BertTokenizer, TFBertForSequenceClassification
from transformers import  TFDistilBertModel, DistilBertTokenizer, TFDistilBertForSequenceClassification
```

```
btokenizer= DistilBertTokenizer.from_pretrained('dbmdz/distilbert-base-turkish-cased')
bmodel = TFDistilBertForSequenceClassification.from_pretrained('dbmdz/distilbert-base-turkish-cased')
```

```
sentences = []
y = []

with open("all_sentiment_train.csv", "r", encoding="utf-8") as file:
    for line in file:
        l = line.strip()
        s, res = l.rsplit("\t", 1)
        sentences.append(s)
        y.append(int(res))
```

```
len(sentences)
```

```
w =[btokenizer.encode(x, add_special_tokens=True, max_length=150, pad_to_max_length=True) for x in sentences]
```

```
max_len = max(len(sent.split()) for sent in sentences)
```

```
max_len
```

```
w[0]
```

```
input_layer = tf.keras.layers.Input(shape = (150,), dtype='int64')

bert = bmodel(input_layer)
bert = bert[0]

#conv=tf.keras.layers.Conv1D(128, kernel_size=5, activation='relu', name="input")(bert)
#pooling = tf.keras.layers.GlobalMaxPooling1D()(bert)
#dense = tf.keras.layers.Dense(10, activation='relu')(bert)

#dense = tf.keras.layers.Dense(5, activation="relu")(bert)
flat = tf.keras.layers.Flatten()(bert)
classifier= tf.keras.layers.Dense(1, activation='sigmoid', name='outputs')(flat)

#dropout = tf.keras.layers.Dropout(0.1)(bert)
#flat = tf.keras.layers.Flatten()(dropout)
#classifier = tf.keras.layers.Dense(units=1)(flat)

keras_model = tf.keras.Model(inputs=input_layer, outputs=classifier)
keras_model.summary()
```

```
keras_model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=3e-05, epsilon=1e-08, clipnorm=1.0), loss="binary_crossentropy", metrics=["accuracy"])
```

```
keras_model.fit(w, y, epochs=5, batch_size=32)
```
